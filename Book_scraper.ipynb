```python
# Import libraries

import requests
from bs4 import BeautifulSoup
import pandas as pd

print("Libraries imported successfully")

```

```python
# Fetch the webpage

# Define the url we want to scrape
url = 'https://books.toscrape.com/'

# Send a GET request to the website
response = requests.get(url)

# Status code of 200 means it is okay
print(f"Status code :{response.status_code}")
```

```python
# Parse the HTML and find the books

soup = BeautifulSoup(response.text, 'html.parser')

# Find all the articles with the class product pod.
books = soup.find_all('article', class_='product_pod')

print(f"Found {len(books)} on this page.")

print('\n--- First 5 books titles ---')
for book in books[:5]:
    title = book.find('h3').find('a')['title']
    print(title)
```

```python
# Extracting more details like price and rating

# Creating an empty list to store all the data
book_data=[]

# Loop through all the books on the page
for book in books:

  # 1. Get the title
  title = book.find('h3').find('a')['title']

  # 2. Get the price
  price = book.find('p', class_='price_color').text

  # 3. Get the rating
  rating = book.find('p', class_='star-rating')['class'][1] # Corrected class name to 'star-rating'

  # 4. Save the info and add it to the list
  book_data.append({ # Corrected typo 'appeand' to 'append'
      'Title': title,
      'Price': price,
      'Rating': rating
  })

# Print all the data
print(f"\nSuccessfully extracted data for {len(book_data)} books!") # Corrected print statement
```

```python
# Convert to pandas and save

df = pd.DataFrame(book_data)

# Display first 5 rows of our scraped data set
print("\n--- Our Scrapped Dataset ---")
display(df.head())

# Save the dataset in .csv so that we can use it later
df.to_csv('scraped_book.csv', index=False)

print("\n Data successfully saved to 'scraped_book.csv'!")
```
